```python
import os
import glob
import pandas as pd
import emoji
def list_csv_files(folder_path):
    # Use glob to find all CSV files in the specified folder
    csv_files = glob.glob(os.path.join(folder_path, "*.csv"))
    return csv_files

def extract_emoji(df):

    emj_tweet=[]

    emj_tweet_dec=[]

    n_emoji=0

    l_emoji=[]

    for i in df['tweets']:

        if len(emoji.emoji_list(str(i)))!=0:

            n_emoji+=1
            emj_tweet.append(i)
            emj_tweet_dec.append(emoji.demojize(i))
            for j in emoji.emoji_list(str(i)):
            
                l_emoji.append(j)

    return [emj_tweet,emj_tweet_dec,l_emoji]

# Example usage
emj_tweet_f=[]
emj_tweet_dec_f=[]
l_emoji_f=[]
folder_path = os.getcwd()
csv_files = list_csv_files(folder_path)
print("CSV files in the folder:")
for file in csv_files:
    df=pd.read_csv(file)
    print(file)
    print(df.columns)
    emj_inf=extract_emoji(df)
    emj_tweet_f.extend(emj_inf[0])
    emj_tweet_dec_f.extend(emj_inf[1])
    l_emoji_f.extend(emj_inf[2])


df_emj_dcd=pd.DataFrame(zip(emj_tweet_f,emj_tweet_dec_f,emj_tweet_rem),columns=["Orig_Tweet","Dec_Tweet","Tweet_wo_emoji"])
```
